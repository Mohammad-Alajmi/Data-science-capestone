```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    r = 6371  # Earth radius in km
    return c * r

# Coordinates of first launch site
launch_site_lat = spacex_df.loc[0, 'Lat']
launch_site_lon = spacex_df.loc[0, 'Long']

# Example coastline coordinates near launch site
coastline_lat = 28.56300
coastline_lon = -80.57000

distance = haversine(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import math
import pandas as pd

# Haversine function to calculate distance in km
def calculate_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    r = 6371  # Earth radius in km
    return c * r

# Example launch site coordinates
launch_site_lat = 28.5623
launch_site_lon = -80.5774

# Example coastline coordinates DataFrame
coastline_coords = pd.DataFrame({
    'Latitude': [28.56367, 28.56400, 28.56000],  # replace/add actual coastline latitudes
    'Longitude': [-80.57163, -80.57200, -80.57000]  # replace/add actual coastline longitudes
})

# Calculate distances from launch site to each coastline point
coastline_coords['Distance_km'] = coastline_coords.apply(
    lambda row: calculate_distance(launch_site_lat, launch_site_lon, row['Latitude'], row['Longitude']), axis=1)

# Find the closest coastline point
closest_point = coastline_coords.loc[coastline_coords['Distance_km'].idxmin()]

print("Closest coastline point:")
print(closest_point)
```

```python
import folium
from folium.features import DivIcon

# Assuming you have:
# closest_point with 'Latitude' and 'Longitude'
# launch_site coordinate: (launch_site_lat, launch_site_lon)
# distance value in kilometers: closest_point['Distance_km']

# Create your base map centered around the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Marker for the launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Marker for the closest coastline point
coastline_coord = [closest_point['Latitude'], closest_point['Longitude']]
distance = closest_point['Distance_km']

distance_marker = folium.Marker(
    location=coastline_coord,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance:.2f} KM</b></div>'
    )
)

distance_marker.add_to(site_map)

# Optional: add a marker for the coastline point itself
folium.Marker(
    location=coastline_coord,
    popup="Closest Coastline",
    icon=folium.Icon(color='green')
).add_to(site_map)

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon

# Example launch site coordinate
launch_site_lat = 28.5623
launch_site_lon = -80.5774
launch_site_coord = [launch_site_lat, launch_site_lon]

# Create base map centered on launch site
site_map = folium.Map(location=launch_site_coord, zoom_start=10)

# Add launch site marker
folium.Marker(
    location=launch_site_coord,
    popup="Launch Site",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Example closest features with coordinates and distances (replace with your data)
features = {
    'Closest City': {'coord': [28.5700, -80.5800], 'distance_km': 1.2},
    'Closest Railway': {'coord': [28.5550, -80.5900], 'distance_km': 2.5},
    'Closest Highway': {'coord': [28.5650, -80.5600], 'distance_km': 3.1}
}

for name, info in features.items():
    coord = info['coord']
    distance = info['distance_km']

    # Add marker with distance label
    distance_marker = folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance:.2f} KM</b></div>'
        ),
        popup=f"{name} - {distance:.2f} KM"
    )
    distance_marker.add_to(site_map)

    # Add a marker icon for the feature
    folium.Marker(
        location=coord,
        popup=name,
        icon=folium.Icon(color='green')
    ).add_to(site_map)

    # Draw a line from launch site to feature
    line = folium.PolyLine(locations=[launch_site_coord, coord], color='purple', weight=2)
    site_map.add_child(line)

# Display the map
site_map
```
